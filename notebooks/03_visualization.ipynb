# Visualization of Income Inequality and Informal Employment in South Africa

This notebook presents a range of visualizations (bar, scatter, box, histograms, etc.) using the cleaned dataset. Each plot is clearly labeled and uses color to highlight trends and patterns. Explanations are provided for each visualization to interpret the results.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
# Load the cleaned and merged dataset
data_path = '../data/data_processed/merged_clean.csv'
df = pd.read_csv(data_path)
print('Loaded data shape:', df.shape)
df.head()
```

## Bar Plot: Average Gini Index and Informal Employment

This bar plot shows the average values of the Gini Index and Informal Employment, allowing for a quick comparison of their central tendencies.

```python
avg_values = df[['Gini_Index', 'Informal_Employment']].mean()
plt.figure(figsize=(7, 5))
sns.barplot(x=avg_values.index, y=avg_values.values, palette=['#1f77b4', '#ff7f0e'])
plt.title('Average Gini Index vs Informal Employment')
plt.ylabel('Average Value')
plt.xlabel('Indicator')
plt.show()
```

## Histogram: Distribution of Gini Index and Informal Employment

Histograms reveal the distribution and spread of each indicator, highlighting skewness and outliers.

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Gini_Index'], bins=10, color='#1f77b4', kde=True)
plt.title('Gini Index Distribution')
plt.xlabel('Gini Index')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.histplot(df['Informal_Employment'], bins=10, color='#ff7f0e', kde=True)
plt.title('Informal Employment Distribution')
plt.xlabel('Informal Employment')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Box Plot: Spread and Outliers

Box plots show the spread and highlight outliers for both indicators.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Gini_Index', 'Informal_Employment']], palette=['#1f77b4', '#ff7f0e'])
plt.title('Boxplot of Gini Index and Informal Employment')
plt.ylabel('Value')
plt.xlabel('Indicator')
plt.show()
```

## Scatter Plot: Relationship Between Indicators

This scatter plot visualizes the relationship between Informal Employment and the Gini Index, with color and labels for clarity.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Informal_Employment', y='Gini_Index', data=df, color='purple', edgecolor='white', s=100)
plt.title('Scatter Plot: Informal Employment vs Gini Index')
plt.xlabel('Informal Employment')
plt.ylabel('Gini Index')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## Trends Over Time: Line Plot

This line plot shows how both indicators change over time, with clear colors and labels.

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Gini_Index'], color='#1f77b4', marker='o', label='Gini Index')
plt.plot(df['Year'], df['Informal_Employment'], color='#ff7f0e', marker='s', label='Informal Employment')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trends in Gini Index and Informal Employment Over Time')
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Interpretation of Trends and Patterns

- The bar plot shows that the average Gini Index is higher than the average informal employment rate, indicating persistent income inequality.
- Histograms and box plots reveal moderate spread and some outliers, with the Gini Index distribution slightly skewed.
- The scatter plot suggests a positive relationship between informal employment and income inequality.
- The line plot shows that both indicators tend to move together over time, with notable increases and decreases during key years.

**Conclusion:**
Clear color schemes, labels, and legends make these visualizations easy to interpret and support the analysis of trends and patterns in South Africa's income inequality and informal economy.

## Advanced Visualizations and Deeper Analysis

we add more plot types, highlight key events, and provide richer explanations for trends and relationships in the data.

```python
# Pairplot for relationships between variables
sns.pairplot(df[['Gini_Index', 'Informal_Employment', 'Year']], diag_kind='kde', palette='husl')
plt.suptitle('Pairplot: Relationships Between Indicators', y=1.02)
plt.show()

# Correlation heatmap
plt.figure(figsize=(6, 4))
corr = df[['Gini_Index', 'Informal_Employment']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Violin plot for distribution shape and outliers
plt.figure(figsize=(8, 5))
sns.violinplot(data=df[['Gini_Index', 'Informal_Employment']], palette=['#1f77b4', '#ff7f0e'])
plt.title('Violin Plot of Gini Index and Informal Employment')
plt.ylabel('Value')
plt.xlabel('Indicator')
plt.show()
```

```python
# Annotated line plot highlighting key years/events
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Gini_Index'], color='#1f77b4', marker='o', label='Gini Index')
plt.plot(df['Year'], df['Informal_Employment'], color='#ff7f0e', marker='s', label='Informal Employment')
key_years = [1994, 2008, 2020]
for year in key_years:
    plt.axvline(x=year, color='red', linestyle=':', alpha=0.7)
    plt.text(year, plt.ylim()[1]*0.95, f'{year}', color='red', ha='center', va='top', fontsize=10, rotation=90)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trends in Gini Index and Informal Employment with Key Events')
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Insights from Advanced Visualizations

- **Pairplot:** Shows the relationships and distributions between Gini Index, Informal Employment, and Year. Diagonal plots reveal distribution shapes, while off-diagonal plots show correlations.
- **Correlation Heatmap:** Quantifies the strength and direction of relationships between indicators. A strong positive value supports the hypothesis of association.
- **Violin Plot:** Combines boxplot and density information, revealing the shape and spread of each indicator, including outliers and multimodal distributions.
- **Annotated Line Plot:** Highlights key years (1994, 2008, 2020) that may correspond to major events in South Africa, helping to contextualize changes in the indicators.

**Conclusion:**
These advanced visualizations provide deeper insights into the data, reveal complex relationships, and make your analysis more comprehensive and visually compelling.
